## Data Preprocessing for ANN

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("BankCustomers.csv")
data
```

```python
geography = pd.get_dummies(data['Geography'],drop_first=True)
```

```python
gender = pd.get_dummies(data['Gender'],drop_first=True)
```

```python
data=pd.concat([data,geography,gender],axis=1)
data
```

```python
data =data.drop(['Geography','Gender','RowNumber','CustomerId','Surname'],axis=1)
```

```python
X = data.drop(['Exited'],axis=1)
X
```

```python
y = data['Exited']
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

### ANN

```python
import keras 
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier = Sequential()
```

```python
classifier.add(Dense(activation='relu',input_dim=11,units=8,kernel_initializer='uniform'))
```

```python
classifier.add(Dense(activation='relu',units=6,kernel_initializer='uniform'))
```

```python
classifier.add(Dense(activation='sigmoid',units=1,kernel_initializer='uniform'))
```

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
classifier.fit(X_train,y_train,batch_size=10,epochs= 100)
```

```python
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm= confusion_matrix(y_test,y_pred)
accuracy = accuracy_score(y_test,y_pred)
accuracy
```
